```python
# train data
from glob import glob
img_li1 = glob('../dataset/yolo_dataset/images/top/*.jpg')
img_li2 = glob('../dataset/yolo_dataset/labels/top/*.txt')
print(len(img_li1))
print(len(img_li2))
```

```python
import os
# yolo 모델 루트 디렉토리 경로
root_dir = "Obj_detection_model/yolov5/"

# 이미지, 라벨 디렉토리 경로
img_dir = '../dataset/yolo_dataset/images'
label_dir = '../dataset/yolo_dataset/labels'

# 상의. 하의 . 하의_상의. 원피스 각각의 이미지 파일을 리스트에 저장해주기
img_top = glob(img_dir + r'/top/*.jpg')
img_pants = glob(img_dir + r'/bottom/*.jpg')
img_tppt = glob(img_dir + r'/top_bottom/*.jpg')
img_onepi = glob(img_dir + r'/onepiece/*.jpg')

# 상의. 하의 . 하의_상의. 원피스 각각의 bbox 파일을 리스트에 저장해주기
label_top = glob(label_dir + r'/top/*.txt')
label_pants = glob(label_dir + r'/bottom/*.txt')
label_tppt = glob(label_dir + r'/top_bottom/*.txt')
label_onepi = glob(label_dir + r'/onepiece/*.txt')
```

```python
## test, val, train 데이터 분리 - 10000장
# train : val = 4 : 1
# top
train_img_top = img_top[:2000]
val_img_top = img_top[2000:2500]
test_img_top = img_top[-500:]
# pants
train_img_pants = img_pants[:2000]
val_img_pants = img_pants[2000:2500]
test_img_pants = img_pants[-500:]
# top_pants
train_img_tppt = img_tppt[:2000]
val_img_tppt = img_tppt[2000:2500]
test_img_tppt = img_tppt[-500:]
# one piece
train_img_onepi = img_onepi[:2000]
val_img_onepi = img_onepi[2000:2500]
test_img_onepi = img_onepi[-500:]

train_img_list = train_img_top + train_img_pants + train_img_tppt + train_img_onepi
val_img_list = val_img_top + val_img_pants + val_img_tppt + val_img_onepi
test_img_list = test_img_top + test_img_pants + test_img_tppt + test_img_onepi
print(len(train_img_list))
print(len(val_img_list))
print(len(test_img_list))
```

```python
## test, val, train 데이터 분리 - 10000장
# train : val = 4 : 1
# top
train_label_top = label_top[:2000]
val_label_top = label_top[2000:2500]
test_label_top = label_top[-500:]
# pants
train_label_pants = label_pants[:2000]
val_label_pants = label_pants[2000:2500]
test_label_pants = label_pants[-500:]
# top_pants
train_label_tppt = label_tppt[:2000]
val_label_tppt = label_tppt[2000:2500]
test_label_tppt = label_tppt[-500:]
# one piece
train_label_onepi = label_onepi[:2000]
val_label_onepi = label_onepi[2000:2500]
test_label_onepi = label_onepi[-500:]

train_label_list = train_label_top + train_label_pants + train_label_tppt + train_label_onepi
val_label_list = val_label_top + val_label_pants + val_label_tppt + val_label_onepi
test_label_list = test_label_top + test_label_pants + test_label_tppt + test_label_onepi
print(len(train_label_list))
print(len(val_label_list))
print(len(test_label_list))
```
